```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Read data from CSV
# Example CSV should have columns: 'Keyword' and 'Frequency'
df = pd.read_csv('/content/word_cloud_p1.csv')

# Convert DataFrame to a dictionary
freq_dict = dict(zip(df['Word'], df['Frequency']))

# Path to the TH-Sarabun-New-Regular font file
font_path = '/content/TH-Sarabun-New-Regular.ttf'

# Create a WordCloud
wordcloud = WordCloud(
    font_path=font_path,  # Thai font
    background_color="white",
    width=1024,
    height=768,
    colormap='plasma',
    scale=3
).generate_from_frequencies(freq_dict)

# Display the WordCloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()

```
