```python
import pandas as pd

df = pd.read_csv('/content/Merge_matching_label.csv')

# Mapping conditions
sdg_mapping = {
    "SDG01_ยุติความยากจน": ["สำนักงานพัฒนาชุมชนจังหวัดปทุมธานี"],
    "SDG02_ยุติความหิวโหย": ["สำนักงานเกษตรและสหกรณ์จังหวัด ปทุมธานี"],
    "SDG03_สุขภาพและความเป็นอยู่ที่ดี": ["สำนักงานสาธารณสุขจังหวัดปทุมธานี"],
    "SDG04_การศึกษาที่มีคุณภาพ ครอบคลุม เท่าเทียม": ["สำนักงานศึกษาธิการจังหวัดปทุมธานี"],
    "SDG05_ความเสมอภาคระหว่างเพศ": ["สำนักงานพัฒนาสังคมและความมั่นคงของมนุษย์ ปทุมธานี"],
    "SDG06_บริหารจัดการน้ำและสุขาภิบาล": ["สำนักงานทรัพยากรธรรมชาติและสิ่งแวดล้อมจังหวัด ปทุมธานี"],
    "SDG07_พลังงานสมัยใหม่ที่ทุกคนเข้าถึงได้": ["สำนักงานพลังงานจังหวัดปทุมธานี"],
    "SDG08_การเติบโตทางเศรษฐกิจและสร้างงานที่มีคุณค่า": ["สำนักงานแรงงานจังหวัดปทุมธานี"],
    "SDG09_โครงสร้างพื้นฐานอุตสาหกรรมและนวัตกรรม": ["สำนักงานอุตสาหกรรมจังหวัดปทุมธานี"],
    "SDG10_ลดความไม่เสมอภาค": ["สำนักงานพัฒนาสังคมและความมั่นคงของมนุษย์ ปทุมธานี"],
    "SDG11_เมืองและถิ่นฐานของมนุษย์ปลอดภัยและยั่งยืน": ["สำนักงานสถิติจังหวัด ปทุมธานี"],
    "SDG12_ผลิตและบริโภคที่ยั่งยืน": ["สำนักงานทรัพยากรธรรมชาติและสิ่งแวดล้อมจังหวัด ปทุมธานี"],
    "SDG13_การรับมือกับการเปลี่ยนแปลงสภาพภูมิอากาศ": ["สำนักงานทรัพยากรธรรมชาติและสิ่งแวดล้อมจังหวัด ปทุมธานี"],
    "SDG15_ปกป้อง ฟื้นฟู ระบบนิเวศน์บนบกอย่างยั่งยืน": ["สำนักงานทรัพยากรธรรมชาติและสิ่งแวดล้อมจังหวัด ปทุมธานี"],
    "SDG16_สังคมสงบสุข ยุติธรรม และมีความรับผิดชอบ": ["สำนักงานยุติธรรม จังหวัดปทุมธานี"],
    "SDG17_กลุ่มหุ้นส่วนเพื่อการพัฒนา": ["สำนักงานจังหวัดปทุมธานี"]
}

# Invert mapping for efficient lookup
inverse_mapping = {}
for sdg, depts in sdg_mapping.items():
    for dept in depts:
        if dept not in inverse_mapping:
            inverse_mapping[dept] = []
        inverse_mapping[dept].append(sdg)

# Function to map SDGs
def map_sdgs(dept_sub_name):
    return ", ".join(inverse_mapping.get(dept_sub_name, ["NOT_relate_to_SDGs"]))

# Apply the mapping
df['relate_to_SDGs'] = df['dept_sub_name'].apply(map_sdgs)

df
```
