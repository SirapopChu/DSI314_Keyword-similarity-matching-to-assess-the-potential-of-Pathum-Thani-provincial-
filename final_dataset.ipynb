```python
import pandas as pd

# ฟังก์ชันแปลงวันที่จากหลายรูปแบบเป็น YYYY-MM-DD
def convert_date(date_str):
    if pd.isna(date_str):  # หากเป็น NaN ให้คืนค่า None
        return None
    try:
        # ลองแปลงรูปแบบที่เป็น dd/MM/YYYY หรือ dd/MM/YY
        if "/" in date_str:
            return pd.to_datetime(date_str, dayfirst=True, errors='coerce').strftime('%Y-%m-%d')

        # ลองแปลงรูปแบบที่เป็น dd เดือน ปี (ภาษาไทย)
        thai_months = {
            "ม.ค.": "01", "ก.พ.": "02", "มี.ค.": "03", "เม.ย.": "04",
            "พ.ค.": "05", "มิ.ย.": "06", "ก.ค.": "07", "ส.ค.": "08",
            "ก.ย.": "09", "ต.ค.": "10", "พ.ย.": "11", "ธ.ค.": "12"
        }
        parts = date_str.split()
        day = int(parts[0])
        month = thai_months.get(parts[1], "01")  # แปลงเดือน
        year = int(parts[2])
        year = year + 2500 if year < 2500 else year  # หากเป็นปีแบบ 60 (สองหลัก) แปลงเป็น พ.ศ.
        year -= 543  # แปลงจาก พ.ศ. เป็น ค.ศ.
        return f"{year:04d}-{month}-{day:02d}"  # รูปแบบ YYYY-MM-DD
    except Exception:
        return None

# แปลงคอลัมน์ _date ทั้งหมด
date_columns = [col for col in df_final.columns if col.endswith('_date')]
for col in date_columns:
    df_final[col] = df_final[col].apply(convert_date)

# ลบคอลัมน์ที่ไม่ต้องการออก
columns_to_drop = ['method_groupname', 'project_location', 'winner_tin', 'contract_no']
df2 = df_final.drop(columns=columns_to_drop, errors='ignore')

# ตรวจสอบผลลัพธ์
df2.head()
```

```python
import pandas as pd

# ตรวจสอบว่ามีค่า NaN ในคอลัมน์ 'year' หรือไม่
# ถ้ามี NaN ให้แทนที่เป็นค่าอื่น เช่น ปีที่พบบ่อยที่สุด หรือกรณีนี้จะใช้การลบ
df2 = df2.dropna(subset=['year'])  # ลบแถวที่มีค่า NaN ในคอลัมน์ 'year'

# แปลงคอลัมน์ 'year' เป็น int
df2['year'] = df2['year'].astype(int)

# แปลงคอลัมน์ 'contract_status' โดยตรง เพื่อแยกสถานะ
df2['status'] = df2['contract_status'].apply(lambda x: 1 if x == 'สิ้นสุดสัญญา' else 0)

# นับจำนวนโครงการทั้งหมดในแต่ละปี
total_projects_by_year = df2.groupby('year').size()

# แยกจำนวนโครงการตามสถานะ 'contract_status' (ระหว่างดำเนินการ, สิ้นสุดสัญญา)
status_projects_by_year = df2.groupby(['year', 'status']).size().unstack(fill_value=0)

# เปลี่ยนชื่อคอลัมน์
status_projects_by_year.columns = ['ระหว่างดำเนินการ', 'สิ้นสุดสัญญา']

# รวมจำนวนโครงการทั้งหมดในแต่ละปี
status_projects_by_year['Total Projects'] = total_projects_by_year

# คำนวณเปอร์เซ็นต์
status_projects_by_year['% ระหว่างดำเนินการ'] = (status_projects_by_year['ระหว่างดำเนินการ'] / status_projects_by_year['Total Projects']) * 100
status_projects_by_year['% สิ้นสุดสัญญา'] = (status_projects_by_year['สิ้นสุดสัญญา'] / status_projects_by_year['Total Projects']) * 100

# แสดงผลลัพธ์
status_projects_by_year
```

```python
import pandas as pd
import plotly.express as px

# ตรวจสอบว่ามีค่า NaN ในคอลัมน์ 'year' หรือไม่
df2 = df2.dropna(subset=['year'])  # ลบแถวที่มีค่า NaN ในคอลัมน์ 'year'

# แปลงคอลัมน์ 'year' เป็น int
df2['year'] = df2['year'].astype(int)

# แปลงคอลัมน์ 'contract_status' โดยตรง เพื่อแยกสถานะ
df2['status'] = df2['contract_status'].apply(lambda x: 'สิ้นสุดสัญญา' if x == 'สิ้นสุดสัญญา' else 'ระหว่างดำเนินการ')

# นับจำนวนโครงการตามปีและสถานะ
status_projects_by_year = df2.groupby(['year', 'status']).size().reset_index(name='count')

# สร้างกราฟแท่งซ้อน
fig = px.bar(status_projects_by_year,
             x='year',
             y='count',
             color='status',
             title='จำนวนโครงการตามสถานะในแต่ละปี',
             labels={'year': 'ปี', 'count': 'จำนวนโครงการ', 'status': 'สถานะโครงการ'},
             color_discrete_map={'ระหว่างดำเนินการ': 'red', 'สิ้นสุดสัญญา': 'green'})

# แสดงกราฟ
fig.show()
```

### โครงการที่ ไม่สอดคล้องกับประเด็นพัฒนา กับ ไม่สอดคล้องตาม SDGs

```python
import pandas as pd

# เช็คจำนวนข้อมูลใน 'strategy_label' ที่ไม่สอดคล้อง
strategy_not_matching = df2[df2['strategy_label'] == 'ไม่สอดคล้องกับประเด็นพัฒนา'].shape[0]
total_strategy = df2.shape[0]
percentage_strategy_not_matching = (strategy_not_matching / total_strategy) * 100

# เช็คจำนวนข้อมูลใน 'relate_to_SDGs' ที่ไม่สอดคล้อง
sdgs_not_matching = df2[df2['relate_to_SDGs'] == 'ไม่สอดคล้องตาม SDGs'].shape[0]
total_sdgs = df2.shape[0]
percentage_sdgs_not_matching = (sdgs_not_matching / total_sdgs) * 100

# สร้าง DataFrame สำหรับสรุปข้อมูล
summary_df = pd.DataFrame({
    'Column': ['strategy_label', 'relate_to_SDGs'],
    'Not Matching Count': [strategy_not_matching, sdgs_not_matching],
    'Total Count': [total_strategy, total_sdgs],
    'Percentage Not Matching': [percentage_strategy_not_matching, percentage_sdgs_not_matching]
})

# แสดงผล
summary_df
```
