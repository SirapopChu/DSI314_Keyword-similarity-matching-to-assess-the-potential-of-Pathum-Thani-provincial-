```python
from pythainlp.wangchanberta import ThaiNameTagger
from pythainlp.tokenize import word_tokenize
from pythainlp.corpus import thai_stopwords
from pythainlp.tag import pos_tag
import pandas as pd
import re

# สร้าง ThaiNameTagger instance
ner = ThaiNameTagger()

# Allow list สำหรับคำที่ต้องการคงไว้ใน final_tokenized
allow_list = {"สะพาน", "คอนกรีตเสริมเหล็ก"}

# ฟังก์ชันสำหรับการแยก Named Entities
def extract_named_entities(text):
    if not text or not isinstance(text, str) or text.strip() == "":
        return [], []

    try:
        # ใช้ get_ner เพื่อดึง Named Entities
        ner_results = ner.get_ner(text)

        # ดึงเฉพาะคำที่เป็น Named Entities (ไม่ใช่ "O")
        named_entities = [(word, tag) for word, tag in ner_results if tag != "O"]

        # ดึงคำอื่นๆ (ที่ tag = "O")
        other_tokens = [word for word, tag in ner_results if tag == "O"]

        return named_entities, other_tokens
    except Exception as e:
        print(f"Error processing text '{text}': {e}")
        return [], []

# ฟังก์ชันกรองคำกริยา, คำคุณศัพท์, คำวิเศษณ์ และคำแปลกๆ
def filter_unwanted_words(words):
    tagged_words = pos_tag(words)
    filtered_words = [
        word for word, tag in tagged_words
        if (word in allow_list) or  # อนุญาตคำใน allow_list
        (not tag.startswith(('V', 'ADJ', 'ADV')) and len(word.strip()) > 1)  # กรองคำ
    ]
    return filtered_words

# ฟังก์ชัน preprocess ข้อมูล
def preprocess_text(text):
    if not text or not isinstance(text, str) or text.strip() == "":
        return [], [], []

    # แก้ไขคำย่อและลบวงเล็บ
    text = text.replace("ค.ส.ล", "คอนกรีตเสริมเหล็ก")
    text = text.replace("ค.ส.ล.", "คอนกรีตเสริมเหล็ก")
    text = text.replace("คสล.", "คอนกรีตเสริมเหล็ก")
    text = text.replace("แอสฟัลท์ติก", "แอสฟัลท์ติก")
    text = re.sub(r'\(.*?\)', '', text)  # ลบเนื้อหาภายในวงเล็บ

    # ดึง Named Entities และ Other Tokens
    named_entities, other_tokens = extract_named_entities(text)

    # กรองคำที่ไม่ต้องการจาก Other Tokens
    filtered_other_tokens = filter_unwanted_words(other_tokens)

    # Tokenize Other Tokens และกรองคำที่ไม่ต้องการอีกรอบ
    tokenized = word_tokenize(' '.join(filtered_other_tokens))
    final_tokenized = [
        word for word in filter_unwanted_words(tokenized)
        if word not in stopwords_set and word.strip()
    ]

    return named_entities, filtered_other_tokens, final_tokenized

# เตรียมชุดข้อมูล
stopwords_set = set(thai_stopwords())

# ใช้ฟังก์ชัน preprocess_text
df_merge[['named_entities', 'other_tokens', 'final_tokenized']] = df_merge['project_name'].apply(
    lambda x: pd.Series(preprocess_text(x))
)

# แสดงผลลัพธ์
print("ผลลัพธ์การทำ NER และ Tokenization:")
df_merge[['project_name', 'named_entities', 'other_tokens', 'final_tokenized']]
```

```python
import re
from pythainlp.corpus import thai_stopwords
from pythainlp.tag import pos_tag

# สร้าง stopwords ที่รวม custom stopwords และคำที่ไม่ต้องการ
stopwords_set = set(thai_stopwords())
exclude_words = {'จํา','กํา', 'ตําบล', "บริเวณ", "ปี", "ปทุมธานี", "ปทุม", "หน้า", "ตารางเมตร", "ศูนย์", "ปุ", "เข", "ตจ", "ํา",
                 "พระ", "ชด", "ริ", "มม.", "สํา", "หรับ", "โครงการ", "โครง", "3.114", "งาน", 'ขับขี่', 'ปลอดภัย', 'วินัย',
                 "ระบบ", "แนว", "ชุด", "คู่", "เมนต์", "เส้น", '""', "เชิง", 'ิกส์', 'ท่องเที่ยว', 'แนวทาง', 'หลา', 'ย่าน', 'กม.',
                 '.45+800', 'เชิงปฏิบัติการ', 'ดูงาน', 'ต้นแบบ', 'หมายเลข', 'ตอน', 'บรรจบ', 'บํา', 'บัด', "ตำบล", '.4', '.8',
                 'จิต', 'อาสา', 'ขุด', 'คัน', 'ศักยภาพ', 'สาร', 'ชีว', 'ภัณฑ์', 'ช่องทาง', 'สิ', 'ค้า', 'แปรรูป', 'เมือง', 'ประชา', 'รัฐ',
                 'อินทรีย์', 'กระบวนการ', 'มูลค่า','ธง', 'ฟ้า', 'ราคา','มูลค่าเพิ่ม','ศักดิ์','มาตร','โอกาส','ถิ่น','เทียว','กรรม','เพณียล',
                 'เกีย','กิจกรรม','เลียบ', 'ซ้าย', 'ขนาด','พื้นที่', 'ตร', 'ราง', 'วี','กิจ', 'กรรมการ','พลัง', 'คน', 'วัย', 'อยู่ดี', 'สุข',
                 'กาย', 'อุ่นใจ', 'โรย','แยกจาก','ชนิด', 'วัตต์', 'กั้น','ศูน์','ศูนย','สาขา','ทางการ','ที่รวม','พื้น',
                 'ช่อง','ลูก','จุด'}

# คำกริยา, คำคุณศัพท์ และคำวิเศษณ์ที่ต้องการกรอง
unwanted_verbs_adj_adv = {"สวย", "งาม", "จัดแสดง", "จำ", "ย่อยสลาย", "ซ่อมแซม", "รณรงค์", "แก้ไขปัญหา", "ยกระดับ", "เพิ่มประสิทธิภาพ", "หน่าย",
                          "สืบสาน", "หลัก", "บูรณาการ", "ยุติธรรม", "ริม", "รักษ์", 'ประโยชน์', "กำ", 'เฝ้า', 'คุณภาพ', 'สี', 'สัมมนา', "ลำ",
                          'ปลอด', 'มาตรฐาน', 'เข้าสู่ระบบ', 'ปราดเปรื่อง', 'ระวัง', 'สะอาด', 'รสชาติ', 'อร่อย', 'เลี้ยงสัตว์', 'ฟื้นฟู', 'เพิ่มปริมาณ',
                          'สูบ', 'ระ', 'สิทธิ', 'พัฒนาการ', 'เกณฑ์', 'อันตราย', 'อย.','ประชาสัมพันธ์','ปราด','อัจฉริยะ','เปรื่อง','เปรือง','สร้างสรรค์', 'แห่งชาติ',
                          'งดงาม','รองรับ','เสริมสร้าง','สามัคคี', 'ปรองดอง','ขยายผล','รัก','ข้าม'}

# คำผิดและคำที่ถูกต้อง
typo_corrections = {
    "เนื่อ": "เนื้อ"
}

# คำที่ต้องการรวม
combine_rules = [
    ["ข้าว", "หอม"],  # รวมเป็น "ข้าวหอม"
    ["ปลา", "น้ำ", "จืด"],  # รวมเป็น "ปลาน้ำจืด"
    ["เนื้อ", "ไก่"] , # รวมเป็น "เนื้อไก่"
    ["กล้วยหอม","ทอง"],
    ["แม่", "น้ำ"],
    ["แหล่ง", "น้ำ"],
    ["ลํา", "คลอง"],
    ['ปลา', 'น้ํา', 'จืด'],
    ['แอ', 'สฟัลท์', 'ติก'],
    ['น้ํา', 'ประปา'],
    ['บ่อ','พัก'],
    ['โซ', 'ล่า', 'เซลล์'],
    ['พันธุกรรม', 'พืช']
]

# ฟังก์ชันแก้คำผิดในโทเคน
def correct_typos_in_tokens(tokens):
    corrected_tokens = []
    for token in tokens:
        if token in typo_corrections:
            corrected_tokens.append(typo_corrections[token])
        else:
            corrected_tokens.append(token)
    return corrected_tokens

# ฟังก์ชันรวมคำตามกฎที่กำหนด
def combine_tokens(tokens):
    i = 0
    combined_tokens = []
    while i < len(tokens):
        matched = False
        for rule in combine_rules:
            # ตรวจสอบว่าคำใน rule ตรงกับตำแหน่งปัจจุบัน
            if tokens[i:i+len(rule)] == rule:
                combined_tokens.append("".join(rule))  # รวมคำ
                i += len(rule)  # ข้ามคำใน rule
                matched = True
                break
        if not matched:
            combined_tokens.append(tokens[i])
            i += 1
    return combined_tokens

# ฟังก์ชันกรองคำที่ไม่ต้องการ
def filter_unwanted_words(final_tokenized):
    # แก้คำผิดในโทเคน
    corrected_tokens = correct_typos_in_tokens(final_tokenized)

    # รวมคำตามกฎที่กำหนด
    combined_tokens = combine_tokens(corrected_tokens)

    # กรองคำที่ไม่ต้องการ
    filtered_words = [
        word for word in combined_tokens
        if not re.match(r'[+\-]?\d+(\.\d+)?([+\-]\d+)*', word)  # กรองตัวเลขทุกรูปแบบ
        and word not in stopwords_set
        and word not in exclude_words
        and word not in unwanted_verbs_adj_adv  # กรองคำกริยา, คำคุณศัพท์, คำวิเศษณ์ที่ไม่ต้องการ
        and not pos_tag([word])[0][1].startswith(('V', 'ADV', 'ADJ'))  # กรองคำกริยา, คำวิเศษณ์, คำคุณศัพท์
        and not re.match(r"^(การ|ความ|จังหวัด|ริมแม่น้ำ|ตำบล|หมู่บ้าน)", word)  # กรองคำที่ขึ้นต้นด้วย "การ" หรือ "ความ"
        and not re.match(r"^\.{1,2}$", word)  # กรองคำที่เป็นจุดหรือเครื่องหมาย
    ]
    return filtered_words

# ใช้ฟังก์ชัน filter_unwanted_words กับ final_tokenized
df_merge['filtered_tokenized'] = df_merge['final_tokenized'].apply(filter_unwanted_words)

# แสดงผลลัพธ์
print("\nCorrected and Combined Tokenized Words")
for index, row in df_merge.iterrows():
    print(f"Project Name: {row['project_name']}")
    print(f"Corrected Words: {row['filtered_tokenized']}\n")
```

```python
from collections import Counter
import math
import pandas as pd

# ฟังก์ชันสำหรับการคำนวณ TF-IDF สำหรับคำนามที่กรองแล้ว
def calculate_tfidf(processed_nouns):
    # รวมทุกคำจาก processed_nouns เป็นลิสต์
    all_words = [word for project in processed_nouns for word in project]

    # คำนวณความถี่คำ (Word Frequency)
    word_counts = Counter(all_words)

    # คำนวณ TF (Term Frequency)
    total_words = len(all_words)
    tf_scores = {word: count / total_words for word, count in word_counts.items()}

    # คำนวณ IDF (Inverse Document Frequency) - สำหรับเอกสารเดียว
    idf_scores = {word: math.log(1 + 1 / 1) for word in word_counts.keys()}  # log(1 + 1/df)

    # คำนวณ TF-IDF
    tfidf_scores = {word: tf_scores[word] * idf_scores[word] for word in word_counts.keys()}

    # สร้าง DataFrame เพื่อแสดงผล
    data = {
        "Word": list(word_counts.keys()),
        "Frequency": list(word_counts.values()),
        "TF": [tf_scores[word] for word in word_counts.keys()],
        "IDF": [idf_scores[word] for word in word_counts.keys()],
        "TF-IDF": [tfidf_scores[word] for word in word_counts.keys()],
    }

    df_tfidf = pd.DataFrame(data).sort_values(by="TF-IDF", ascending=False).reset_index(drop=True)

    return df_tfidf

# คำนวณ TF-IDF สำหรับข้อมูลใน 'tokenized_other' ของ df
df_tfidf = calculate_tfidf(df_merge['filtered_tokenized'])

# แสดงผล Top 10 คำที่มี TF-IDF สูงสุด
print("\nTop 10 Words by TF-IDF:")
df_tfidf
```

```python
from collections import Counter
import math
import pandas as pd

# ฟังก์ชันสำหรับการคำนวณ TF-IDF สำหรับคำนามที่กรองแล้ว
def calculate_tfidf(processed_nouns):
    # รวมทุกคำจาก processed_nouns เป็นลิสต์
    all_words = [word for project in processed_nouns for word in project]

    # คำนวณความถี่คำ (Word Frequency)
    word_counts = Counter(all_words)

    # คำนวณ TF (Term Frequency)
    total_words = len(all_words)
    tf_scores = {word: count / total_words for word, count in word_counts.items()}

    # คำนวณ IDF (Inverse Document Frequency) - สำหรับเอกสารเดียว
    idf_scores = {word: math.log(1 + 1 / 1) for word in word_counts.keys()}  # log(1 + 1/df)

    # คำนวณ TF-IDF
    tfidf_scores = {word: tf_scores[word] * idf_scores[word] for word in word_counts.keys()}

    # สร้าง DataFrame เพื่อแสดงผล
    data = {
        "Word": list(word_counts.keys()),
        "Frequency": list(word_counts.values()),
        "TF": [tf_scores[word] for word in word_counts.keys()],
        "IDF": [idf_scores[word] for word in word_counts.keys()],
        "TF-IDF": [tfidf_scores[word] for word in word_counts.keys()],
    }

    df_tfidf_1 = pd.DataFrame(data).sort_values(by="TF-IDF", ascending=False).reset_index(drop=True)

    return df_tfidf_1

# คำนวณ TF-IDF สำหรับข้อมูลใน 'tokenized_other' ของ df
df_tfidf_1 = calculate_tfidf(issue_1['filtered_tokenized'])

# แสดงผล Top 10 คำที่มี TF-IDF สูงสุด
print("\nTop 10 Words by TF-IDF:")
df_tfidf_1.head(10)
```

```python
from pythainlp.wangchanberta import ThaiNameTagger
from pythainlp.tokenize import word_tokenize
from pythainlp.corpus import thai_stopwords
from pythainlp.tag import pos_tag
import pandas as pd
import re

# สร้าง ThaiNameTagger instance
ner = ThaiNameTagger()

# Allow list สำหรับคำที่ต้องการคงไว้ใน final_tokenized
allow_list = {"สะพาน", "คอนกรีตเสริมเหล็ก"}

# ฟังก์ชันสำหรับการแยก Named Entities
def extract_named_entities(text):
    if not text or not isinstance(text, str) or text.strip() == "":
        return [], []

    try:
        # ใช้ get_ner เพื่อดึง Named Entities
        ner_results = ner.get_ner(text)

        # ดึงเฉพาะคำที่เป็น Named Entities (ไม่ใช่ "O")
        named_entities = [(word, tag) for word, tag in ner_results if tag != "O"]

        # ดึงคำอื่นๆ (ที่ tag = "O")
        other_tokens = [word for word, tag in ner_results if tag == "O"]

        return named_entities, other_tokens
    except Exception as e:
        print(f"Error processing text '{text}': {e}")
        return [], []

# ฟังก์ชันกรองคำกริยา, คำคุณศัพท์, คำวิเศษณ์ และคำแปลกๆ
def filter_unwanted_words(words):
    tagged_words = pos_tag(words)
    filtered_words = [
        word for word, tag in tagged_words
        if (word in allow_list) or  # อนุญาตคำใน allow_list
        (not tag.startswith(('V', 'ADJ', 'ADV')) and len(word.strip()) > 1)  # กรองคำ
    ]
    return filtered_words

# ฟังก์ชัน preprocess ข้อมูล
def preprocess_text(text):
    if not text or not isinstance(text, str) or text.strip() == "":
        return [], [], []

    # แก้ไขคำย่อและลบวงเล็บ
    text = text.replace("ค.ส.ล", "คอนกรีตเสริมเหล็ก")
    text = text.replace("ค.ส.ล.", "คอนกรีตเสริมเหล็ก")
    text = text.replace("คสล.", "คอนกรีตเสริมเหล็ก")
    text = text.replace("แอสฟัลท์ติก", "แอสฟัลท์ติก")
    text = re.sub(r'\(.*?\)', '', text)  # ลบเนื้อหาภายในวงเล็บ

    # ดึง Named Entities และ Other Tokens
    named_entities, other_tokens = extract_named_entities(text)

    # กรองคำที่ไม่ต้องการจาก Other Tokens
    filtered_other_tokens = filter_unwanted_words(other_tokens)

    # Tokenize Other Tokens และกรองคำที่ไม่ต้องการอีกรอบ
    tokenized = word_tokenize(' '.join(filtered_other_tokens))
    final_tokenized = [
        word for word in filter_unwanted_words(tokenized)
        if word not in stopwords_set and word.strip()
    ]

    return named_entities, filtered_other_tokens, final_tokenized

# เตรียมชุดข้อมูล
stopwords_set = set(thai_stopwords())

# ใช้ฟังก์ชัน preprocess_text
issue_4[['named_entities', 'other_tokens', 'final_tokenized']] = issue_4['project_name'].apply(
    lambda x: pd.Series(preprocess_text(x))
)

# แสดงผลลัพธ์
print("ผลลัพธ์การทำ NER และ Tokenization:")
issue_4[['project_name', 'named_entities', 'other_tokens', 'final_tokenized']]
```

```python
import re
from pythainlp.corpus import thai_stopwords
from pythainlp.tag import pos_tag

# สร้าง stopwords ที่รวม custom stopwords และคำที่ไม่ต้องการ
stopwords_set = set(thai_stopwords())
exclude_words = {'จํา','กํา', 'ตําบล', "บริเวณ", "ปี", "ปทุมธานี", "ปทุม", "หน้า", "ตารางเมตร", "ศูนย์", "ปุ", "เข", "ตจ", "ํา",
                 "พระ", "ชด", "ริ", "มม.", "สํา", "หรับ", "โครงการ", "โครง", "3.114", "งาน", 'ขับขี่', 'ปลอดภัย', 'วินัย',
                 "ระบบ", "แนว", "ชุด", "คู่", "เมนต์", "เส้น", '""', "เชิง", 'ิกส์', 'ท่องเที่ยว', 'แนวทาง', 'หลา', 'ย่าน', 'กม.',
                 '.45+800', 'เชิงปฏิบัติการ', 'ดูงาน', 'ต้นแบบ', 'หมายเลข', 'ตอน', 'บรรจบ', 'บํา', 'บัด', "ตำบล", '.4', '.8',
                 'จิต', 'อาสา', 'ขุด', 'คัน', 'ศักยภาพ', 'สาร', 'ชีว', 'ภัณฑ์', 'ช่องทาง', 'สิ', 'ค้า', 'แปรรูป', 'เมือง', 'ประชา', 'รัฐ',
                 'อินทรีย์', 'กระบวนการ', 'มูลค่า','ธง', 'ฟ้า', 'ราคา','มูลค่าเพิ่ม','ศักดิ์','มาตร','โอกาส','ถิ่น','เทียว','กรรม','เพณียล',
                 'เกีย','กิจกรรม','เลียบ', 'ซ้าย', 'ขนาด','พื้นที่', 'ตร', 'ราง', 'วี','กิจ', 'กรรมการ','พลัง', 'คน', 'วัย', 'อยู่ดี', 'สุข',
                 'กาย', 'อุ่นใจ', 'โรย','แยกจาก','ชนิด', 'วัตต์', 'กั้น','ศูน์','ศูนย','สาขา','ทางการ','ที่รวม','พื้น',
                 'ช่อง','ลูก','จุด'}

# คำกริยา, คำคุณศัพท์ และคำวิเศษณ์ที่ต้องการกรอง
unwanted_verbs_adj_adv = {"สวย", "งาม", "จัดแสดง", "จำ", "ย่อยสลาย", "ซ่อมแซม", "รณรงค์", "แก้ไขปัญหา", "ยกระดับ", "เพิ่มประสิทธิภาพ", "หน่าย",
                          "สืบสาน", "หลัก", "บูรณาการ", "ยุติธรรม", "ริม", "รักษ์", 'ประโยชน์', "กำ", 'เฝ้า', 'คุณภาพ', 'สี', 'สัมมนา', "ลำ",
                          'ปลอด', 'มาตรฐาน', 'เข้าสู่ระบบ', 'ปราดเปรื่อง', 'ระวัง', 'สะอาด', 'รสชาติ', 'อร่อย', 'เลี้ยงสัตว์', 'ฟื้นฟู', 'เพิ่มปริมาณ',
                          'สูบ', 'ระ', 'สิทธิ', 'พัฒนาการ', 'เกณฑ์', 'อันตราย', 'อย.','ประชาสัมพันธ์','ปราด','อัจฉริยะ','เปรื่อง','เปรือง','สร้างสรรค์', 'แห่งชาติ',
                          'งดงาม','รองรับ','เสริมสร้าง','สามัคคี', 'ปรองดอง','ขยายผล','รัก','ข้าม'}

# คำผิดและคำที่ถูกต้อง
typo_corrections = {
    "เนื่อ": "เนื้อ"
}

# คำที่ต้องการรวม
combine_rules = [
    ["ข้าว", "หอม"],  # รวมเป็น "ข้าวหอม"
    ["ปลา", "น้ำ", "จืด"],  # รวมเป็น "ปลาน้ำจืด"
    ["เนื้อ", "ไก่"] , # รวมเป็น "เนื้อไก่"
    ["กล้วยหอม","ทอง"],
    ["แม่", "น้ำ"],
    ["แหล่ง", "น้ำ"],
    ["ลํา", "คลอง"],
    ['ปลา', 'น้ํา', 'จืด'],
    ['แอ', 'สฟัลท์', 'ติก'],
    ['น้ํา', 'ประปา'],
    ['บ่อ','พัก'],
    ['โซ', 'ล่า', 'เซลล์'],
    ['พันธุกรรม', 'พืช']
]

# ฟังก์ชันแก้คำผิดในโทเคน
def correct_typos_in_tokens(tokens):
    corrected_tokens = []
    for token in tokens:
        if token in typo_corrections:
            corrected_tokens.append(typo_corrections[token])
        else:
            corrected_tokens.append(token)
    return corrected_tokens

# ฟังก์ชันรวมคำตามกฎที่กำหนด
def combine_tokens(tokens):
    i = 0
    combined_tokens = []
    while i < len(tokens):
        matched = False
        for rule in combine_rules:
            # ตรวจสอบว่าคำใน rule ตรงกับตำแหน่งปัจจุบัน
            if tokens[i:i+len(rule)] == rule:
                combined_tokens.append("".join(rule))  # รวมคำ
                i += len(rule)  # ข้ามคำใน rule
                matched = True
                break
        if not matched:
            combined_tokens.append(tokens[i])
            i += 1
    return combined_tokens

# ฟังก์ชันกรองคำที่ไม่ต้องการ
def filter_unwanted_words(final_tokenized):
    # แก้คำผิดในโทเคน
    corrected_tokens = correct_typos_in_tokens(final_tokenized)

    # รวมคำตามกฎที่กำหนด
    combined_tokens = combine_tokens(corrected_tokens)

    # กรองคำที่ไม่ต้องการ
    filtered_words = [
        word for word in combined_tokens
        if not re.match(r'[+\-]?\d+(\.\d+)?([+\-]\d+)*', word)  # กรองตัวเลขทุกรูปแบบ
        and word not in stopwords_set
        and word not in exclude_words
        and word not in unwanted_verbs_adj_adv  # กรองคำกริยา, คำคุณศัพท์, คำวิเศษณ์ที่ไม่ต้องการ
        and not pos_tag([word])[0][1].startswith(('V', 'ADV', 'ADJ'))  # กรองคำกริยา, คำวิเศษณ์, คำคุณศัพท์
        and not re.match(r"^(การ|ความ|จังหวัด|ริมแม่น้ำ|ตำบล|หมู่บ้าน)", word)  # กรองคำที่ขึ้นต้นด้วย "การ" หรือ "ความ"
        and not re.match(r"^\.{1,2}$", word)  # กรองคำที่เป็นจุดหรือเครื่องหมาย
    ]
    return filtered_words

# ใช้ฟังก์ชัน filter_unwanted_words กับ final_tokenized
issue_4['filtered_tokenized'] = issue_4['final_tokenized'].apply(filter_unwanted_words)

# แสดงผลลัพธ์
print("\nCorrected and Combined Tokenized Words")
for index, row in issue_4.iterrows():
    print(f"Project Name: {row['project_name']}")
    print(f"Corrected Words: {row['filtered_tokenized']}\n")
```

```python
from collections import Counter
import math
import pandas as pd

# ฟังก์ชันสำหรับการคำนวณ TF-IDF สำหรับคำนามที่กรองแล้ว
def calculate_tfidf(processed_nouns):
    # รวมทุกคำจาก processed_nouns เป็นลิสต์
    all_words = [word for project in processed_nouns for word in project]

    # คำนวณความถี่คำ (Word Frequency)
    word_counts = Counter(all_words)

    # คำนวณ TF (Term Frequency)
    total_words = len(all_words)
    tf_scores = {word: count / total_words for word, count in word_counts.items()}

    # คำนวณ IDF (Inverse Document Frequency) - สำหรับเอกสารเดียว
    idf_scores = {word: math.log(1 + 1 / 1) for word in word_counts.keys()}  # log(1 + 1/df)

    # คำนวณ TF-IDF
    tfidf_scores = {word: tf_scores[word] * idf_scores[word] for word in word_counts.keys()}

    # สร้าง DataFrame เพื่อแสดงผล
    data = {
        "Word": list(word_counts.keys()),
        "Frequency": list(word_counts.values()),
        "TF": [tf_scores[word] for word in word_counts.keys()],
        "IDF": [idf_scores[word] for word in word_counts.keys()],
        "TF-IDF": [tfidf_scores[word] for word in word_counts.keys()],
    }

    df_tfidf_4 = pd.DataFrame(data).sort_values(by="TF-IDF", ascending=False).reset_index(drop=True)

    return df_tfidf_4

# คำนวณ TF-IDF สำหรับข้อมูลใน 'tokenized_other' ของ df
df_tfidf_4 = calculate_tfidf(issue_4['filtered_tokenized'])

# แสดงผล Top 10 คำที่มี TF-IDF สูงสุด
print("\nTop 10 Words by TF-IDF:")
df_tfidf_4.head(10)
```
